```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
np.random.seed(0)
```

```python
df = pd.read_excel('7Mayis.xlsx', 
                 parse_dates={'dt' : ['Tarih', 'Saat']}, 
                  index_col='dt')
```

```python
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
##checking correlation
```

```python
corr.sort_values('GIP AOF (TL/MWh)',ascending=False)
```

```python
y = df["GIP AOF (TL/MWh)"] 
df.drop('GIP AOF (TL/MWh)', axis = 1, inplace = True)
##Set the target value and drop from data
```

```python
a = ["PTF (TL/MWh)","GOP lem Hacmi (TL)","Toplam Uretim (MWh)","PTF (h-1)","GIP (h-1)","PTF (h-2)","GIP (h-2)",
     "PTF (h-3)","GIP (h-3)","Demand","PTF (h-24)","GIP (h-24)","PTF (h-168)","GIP (h-168)"]
```

```python
x = df[a]
x.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = .30, shuffle = False)
X_train
###train and test split
```

```python
X_train.columns
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler() 
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train= pd.DataFrame(X_train, columns= a)
X_test = pd.DataFrame(X_test, columns= a )
X_train
# Standardization
```

```python
from mlxtend.feature_selection import ExhaustiveFeatureSelector as EFS
# Create lienar Regression classifier
lmodel = LinearRegression()
efs = EFS(lmodel, 
           min_features=1,
           max_features=9,
           scoring='neg_mean_squared_error',
           print_progress=True,
           cv=5)

feature_names = (a)
efs = efs.fit(X_train, y_train, custom_feature_names=feature_names)

print('Best subset (indices):', efs.best_idx_)
print('Best subset (corresponding names):', efs.best_feature_names_)
```

```python
pd.DataFrame.from_dict(efs.get_metric_dict()).T.sort_values('avg_score',ascending=False)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
X_train_selected=X_train.iloc[:,[0, 3, 4, 5, 6, 10, 11, 12, 13]]
X_test_selected=X_test.iloc[:,[0, 3, 4, 5, 6, 10, 11, 12, 13]]
### creating new dataset with selected features
lmodel=LinearRegression()
lmodel.fit(X_train_selected, y_train)
y_pred=lmodel.predict(X_test_selected)
mae=mean_absolute_error(y_test,y_pred)
y_pred_train=lmodel.predict(X_train_selected)
mae_train=mean_absolute_error(y_train,y_pred_train)
X_train_selected
print(f"test mae {mae}")
print(f"train mae {mae_train}")
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
lmodel=LinearRegression()
sfs = SFS(lmodel, 
           k_features=9, 
           forward=True, 
           floating=False, 
           verbose=2,
           scoring='neg_mean_squared_error',
           cv=5)
feature_names = a
sfs = sfs.fit(X_train, y_train, custom_feature_names=feature_names)
sfs.subsets_




print('Best subset (corresponding names):', sfs.best_feature_names_)
```

```python
pd.DataFrame.from_dict(sfs.get_metric_dict()).T.sort_values('avg_score',ascending=False)
##we have the same result as previous one so we will have the same result 
```

```python
from sklearn.linear_model import Ridge, RidgeCV, Lasso, LassoCV
###
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
lambdas = np.logspace(-4,-0.5,60)
scoresCV = []
for l in lambdas:
    RidgeReg = Ridge(alpha=l)
    RidgeReg.fit(X_train, y_train)    
    scoreCV = cross_val_score(RidgeReg, X_train, y_train, scoring='neg_mean_squared_error',
                             cv=KFold(n_splits=10, shuffle=True,
                                            random_state=1))
    scoresCV.append([l,-1*np.mean(scoreCV)])
df = pd.DataFrame(scoresCV,columns=['Lambda','Validation Error'])
df
```

```python
df.sort_values(by=['Validation Error'])
```

```python
ridge = Ridge(alpha = 0.316228)
ridge.fit(X_train, y_train)     # Fit a ridge regression on the training data        
pred = ridge.predict(X_test)           # Use this model to predict the test data 
print(pd.Series(ridge.coef_, index = x.columns)) # Print coefficients
print(mean_absolute_error(y_test, pred))   # Calculate the test MAE
```

```python
alphas = np.logspace(-4,-0.5,60)
ridgecv = RidgeCV(alphas = alphas,cv = 10, scoring = 'neg_mean_squared_error')
ridgecv.fit(X_train, y_train)
ridgecv.alpha_
```

```python
ridge = Ridge(alpha = ridgecv.alpha_)
ridge.fit(X_train, y_train)             
pred = ridge.predict(X_test)  
trainpred= ridge.predict(X_train)
print(pd.Series(ridge.coef_, index = x.columns)) 
print(f"mae {mean_absolute_error(y_test, pred)}")   
print(f"train mae {mean_absolute_error(y_train, trainpred)}")   
```
